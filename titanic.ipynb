## Selecting a subset of a dataframe

```python
passanger_ages = df[["Age", "Fare"]]
passanger_ages.shape
```

## Filtering rows 

```python
above_50 = df[df["Age"] > 50]
above_50.head()
```

## Titanic passengers from cabin class 2 and 3

```python
# passengers_23 = df[(df["Pclass"] == 2) | (df["Pclass"] == 3)] or write like the below line. They both yield the same results
passengers_23 = df[df["Pclass"].isin([2,3])]
passengers_23.head()
```

## Find non null values for a specific series

```python
sex_no_na = df[df["Sex"].notna()]
sex_no_na.shape
```

## Filter with both, rows and columns with loc operator

```python
#Get names who spent more than 7000
fare_more_than_7000 = df.loc[df["Fare"] > 7, "Name"]
fare_more_than_7000
```

```python
#Filter based on the index of rows and columns using iloc
#get the first 5 rows and first 3 columns
filter_data = df.iloc[0:5, 0:3]
filter_data
```

```python
#Assign new values to the data we filter using loc/ iloc
df.iloc[0:2,5] = "Idiots"
df.head(5)
```
